```r
library(purrr)
library(hash)
library(rlist)
```

```r
changes <- read.table('./react.txt', header=TRUE)
changes$Total.Diff <- changes$Lines.Added + changes$Lines.Deleted
changes
```

```r
expected_month <- c(
    'May-2019:June-2019',
    'June-2019:July-2019',
    'July-2019:Aug-2019',
    'Aug-2019:Sept-2019',
    'Sept-2019:Oct-2019',
    'Oct-2019:Nov-2019',
    'Nov-2019:Dec-2019',
    'Dec-2019:Jan-2020',
    'Jan-2020:Feb-2020',
    'Feb-2020:March-2020',
    'March-2020:April-2020',
    'April-2020:May-2020'
)

backfill <- function(vector, month, current) {
    if (expected_month[[current]] != month) {
        index_of_expected_month <- match(month, expected_month)
        if (index_of_expected_month > current) {
            num_to_backfill <- index_of_expected_month - current
            for (i in 1:num_to_backfill) {
                vector <- list.append(vector, 0)
            }
        }
    }
    vector
}

pad <- function(a_list) {
    if (length(a_list) < 12) {
        zeros_to_add <- (12 - length(a_list))
        for (i in 1:zeros_to_add) {
            a_list <- list.append(a_list, 0)
        }        
    }
    a_list
}

files <- changes$Path
# Reduce data frame to mapping of files to month-diff pairs
hash <- hash()
for (i in 1:length(files)) {
    file <- as.character(files[[i]])
    if (!has.key(file, hash)) {
        hash[[file]] <- list()
    }
    month <- as.character(changes[[1]][[i]])
    diff <- changes[[5]][[i]]
    updated_list <- list.append(hash[[file]], list(month, diff)) 
    hash[[file]] <- updated_list
}

# Iterate through keys of map and build time-series vectors for each file
files_to_time_series <- hash()
for (k in keys(hash)) {
    entries <- hash[[k]]
    a_list <- list()
    month_index <- 1
    for (i in 1:length(entries)) {
        entry <- entries[[i]]
        month <- entry[[1]]        
        diff <- entry[[2]]
        old_length <- length(a_list)
        a_list <- backfill(a_list, month, month_index)
        new_length <- length(a_list)
        month_index <- month_index + (new_length - old_length) + 1
        a_list <- list.append(a_list, diff)
    }
    files_to_time_series[[k]] <- pad(a_list)
}

files_to_time_series

# calculate the auto correlation for each vector
auto_correlations <- list()
for (k in keys(files_to_time_series)) {
    diff_time_series <- unlist(files_to_time_series[[k]])
    auto_corr <- acf(diff_time_series, lag.max = 1, plot = FALSE)$acf    
    auto_correlations <- list.append(auto_correlations, auto_corr)
}

# plot data
acf_matrix <- t(array(unlist(auto_correlations), dim=c(2, length(auto_correlations))))
month_lag <- apply(acf_matrix, c(2), hist)[[2]]
month_lag
# length(month_lag$out)
```

```r
files_to_quarterly_time_series <- hash()
for (k in keys(files_to_time_series)) {
    monthly_time_series_as_vector <- unlist(files_to_time_series[[k]])
    # convert to real time-series so we can easily use aggregate
    monthly_time_series <- ts(monthly_time_series_as_vector, frequency=12)
    files_to_quarterly_time_series[[k]] <- unclass(aggregate(monthly_time_series, nfrequency=4))    
}

auto_correlations <- list()
for (k in keys(files_to_quarterly_time_series)) {
    diff_time_series <- unlist(files_to_quarterly_time_series[[k]])
    auto_corr <- acf(diff_time_series, lag.max = 1, plot = FALSE)$acf    
    auto_correlations <- list.append(auto_correlations, auto_corr)
}

# plot data
acf_matrix <- t(array(unlist(auto_correlations), dim=c(2, length(auto_correlations))))
quarterly_lag <- apply(acf_matrix, c(2), hist)[[2]]
quarterly_lag
# length(quarterly_lag$out)
```
